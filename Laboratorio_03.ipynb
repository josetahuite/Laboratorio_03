1.  ¿Cuántos Municipios tienen más de 100,000 habitantes?

```python

pob_mun = pd.DataFrame(data.groupby('Municipio').Poblacion.sum())
pob_mun_100k = pob_mun[pob_mun.Poblacion > 100000]
valor = len(pob_mun_100k)
print(f"la cantidad de municipios con más de 100k personas es de {valor}")
```
